```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")

def clean_data(df):
    del df['Ticket']
    del df['Name']
    del df['SibSp']
    del df['Parch']
    del df['Cabin']
    del df['Embarked']
    del df['Fare']

    df['Sex'] = df['Sex'].replace('male', 0).replace('female', 1)
    df = df.rename(columns = {'Sex': 'IsFemale'})

clean_data(df)
df.iloc[0:20]
```

```python
# Linear regression to fill in missing ages
from sklearn import linear_model

non_age_columns = ['Pclass', 'IsFemale']

complete_dataset = df[df['Age'] > 0]
X = complete_dataset[non_age_columns]
Y = complete_dataset[['Age']]
age_regressor = linear_model.LinearRegression()
age_regressor.fit(X, Y)

for index, row in df.iterrows():
    if np.isnan(row['Age']):
        estimated_age = age_regressor.predict(row[non_age_columns].reshape(1, -1))[0][0]
        df.ix[index, 'Age'] = estimated_age
        
df.iloc[0:20]
```

```python
# Train a neural network to estimate survival

from sklearn.neural_network import MLPClassifier

non_survival_columns = ['Pclass', 'IsFemale', 'Age']

X = df[non_survival_columns]
y = df.as_matrix(['Survived']).T[0]

survival_classifier = MLPClassifier(solver='lbfgs', alpha=0.001, hidden_layer_sizes=(10, 10), random_state=1)
survival_classifier.fit(X, y)
print(survival_classifier.score(X, y))
```

```python
# Make estimations

test_data = pd.read_csv("test.csv")
test_data['Survived'] = 0 # Default estimation
clean_data(test_data)
test_data = test_data.rename(columns = {'Sex': 'IsFemale'})

for index, row in test_data.iterrows():
    print(index)
    print(row[non_survival_columns])
    if (np.isnan(row[''])):
        test_data.ix[index, 'Age']
    test_data.ix[index, 'Survived'] = survival_classifier.predict(row[non_survival_columns])
    
1
```
